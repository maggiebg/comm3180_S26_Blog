# Philadelphia Weather on Valentine's Day from the Past 20 Years
### Records from 2005-2025

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
weather_df = pd.read_csv('data/philadelphia_weather_2005_to_2025.csv')
weather_df = weather_df.assign(date=pd.to_datetime(weather_df['date']))
```

```python
valentines_weather = weather_df[ (weather_df['date'].dt.month == 2) & (weather_df['date'].dt.day == 14) ]
valentines_weather
```

```python
valentines_weather['high']
valentines_high = valentines_weather['high']
valentines_high
```

```python
valentines_high['date'] = pd.to_datetime(valentines_high['date'])
valentines_high['date']
```

```python
valentines_high['year'] = valentines_high['date'].dt.year
valentines_high['year']
```

```python
valentines_weather.query("day=='February 14'")
```

```python
valentines_weather.query("day=='February 14'").set_index('year')[['high','low']]
val_high_low = valentines_weather.query("day=='February 14'").set_index('year')[['high','low']]
val_high_low
```

```python
val_high_low.plot(kind='line', style='-o')
plt.title('Philadelphia High and Low temperatures on Valentines Day 2005-2025')
plt.show()
```

```python
warmest_valentines = valentines_weather['high'].max()
warmest_valentines
```

```python
max_idx = valentines_weather['high'].idxmax()
valentines_weather.loc[max_idx]
```

The highest temperature over the past 20 years on Valentine's day was in 2011, reaching 62 degrees.

```python
coldest_valentines = valentines_weather['low'].min()
coldest_valentines
```

```python
min_idx = valentines_weather['low'].idxmin()
valentines_weather.loc[min_idx]
```

The lowest temperature over the past 20 years on Valentine's day was in 2016, reaching only 20 degrees.

```python
is_rain = valentines_weather['rain']>0
is_rain.sum()
```

There were 5 Valentine's days in the past 20 years where there was rain.

```python
prob_rain = (valentines_weather['rain']>0).mean()
prob_rain*100
```

```python
print(f"The probability of rain on Valentine's Day: {prob_rain:.23%}")
```

```python
is_snow = valentines_weather['snow']>0
is_snow.sum()
```

There were 3 Valentine's days in the past 20 years where there was snow.

```python
prob_snow = (valentines_weather['snow']>0).mean()
prob_snow*100
```

```python
print(f"The probability of snow on Valentine's Day: {prob_snow:.14%}")
```
